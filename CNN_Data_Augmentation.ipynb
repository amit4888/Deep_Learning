```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img("/content/Screenshot 2024-06-10 025649.png",target_size=(1500,1500))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,1500,1500,3)
```

```python
input_batch
```

```python
i = 0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='/content/drive/MyDrive/Train'):
  i+=1
  if i>10:
    break
```
